# Map Specification

The following provides illustrated examples of how to construct your coursework map at home using the black vinyl tape (electrical tape) provided in your kit, and all applicable rules.

<br><br><br><br>

## Map Components & Rules

To secure all possible points in Assessment 1, your map must have all of the map components detailed in this specification.  These must also be applied with the following general rules.  Example correct and incorrect maps are provided at the tail of this document.

### General Rules:
1. All straight line segments must be a minimum of 20cm long.  
2. Your  map must not have any lines of symmetry (e.g., the map cannot be the same route when travelled forwards and backwards).  
3. You may add further line segments to best make use of the physical space available to you.
4. You cannot reduce line segments to 0cm, to bunch up components A,B,C,D (see map examples below).
5. When combining map components (A, B, C, D), you may reduce joining straight line segments down to a minimum of 20cm length.
6. All map components must be placed ahead of the `Starting Area`.  No map components may go behind a starting area.
7. The line directly in front of the starting area must be at 90 degrees (perpendicular) to the direction of travel out of the starting area.  
8. The line directly in front of the starting area must be a minimum of 10cm away from the end of the starting area.
9. You must place a second starting area at the opposite end of your line, such that your Romi can attempt the course back in the other direction.
10. Starting areas must each be placed at the start / terminus of the line, and not somewhere along the lines middle length.
11. Your second start area must be placed at the other line terminus, e.g., they cannot both lead the Romi onto the same starting point of the line map.  
12. You may add more than components A, B, C, & D to your map, and create your own challenges if you wish.  This will not secure you extra marks, and should be taken as a matter of interest or convenience.
13. You are not able to secure any further marks or lenience in assessment by creating a line which is more difficult than necessary.
14. If you have any doubt or concern you should contact a member of teaching staff.  There will be no lenience in assessment to compensate for negligence.  


### Map Component Starting Area:

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/start_area.png?raw=true"/>

You must construct two start area as illustrated above (not to scale), adhering to the following rules:
- It must have a depth (measured internal to the lines) of a **minimum** of 30cm.  Shorter starting areas are not permitted.
- It must have a width (measured internal to the lines) of a **maximum** of 18cm.  Wider starting areas are not permitted.
- Your start box must end a mimumum of 10cm away from the line course itself.
- Your two start box must be placed at either end of your line challenge, such that your Romi can attempt to follow the line in either direction.

### Map Component A:

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/map_component_a.png?raw=true"/>

Your map should have a minimum of one curve with the following parameters:
- The curve should have a radius not smaller than 15cm (the diameter of the Romi).
- The curve should have a total rotation not less than 180 degrees.

You may add more than one curve if you wish, with other curves not needing to adhere to the above rules.  It is recommended you use something like a dinner plate as a guide to lay your curve with tape (video example).

Your map should include two 45 degree corners and two 90 degree corners, such that following a direction of a travel, your Romi will need to take a right-hand and left-hand turn on both components B & C during one complete line traversal.
- You can use the corner of a page of A4 paper for a guide to 90 degrees.
- You can fold over a corner of an A4 paper for a guide to 45 degrees.

### Map Component D:

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/map_component_d.png?raw=true"/>

Your map should include a short break in the line of a distance of 4cm minimum, adhering to the following rules:
- You may place this break any where along your line, so long as it remains within the expected travel of your Romi along the complete line length.
- You may have a line break larger than 4cm, but not smaller.
- Your line break does not need to be in a consecutive line, however it should be noted that a line break on a corner makes this component significantly harder.  There are no further marks available for a difficult placement  of a line break.

## Example Incorrect Maps:

### No convenient Join-Line section:

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/no_straight_leadon.png?raw=true"/>

The above incorrect map illustrates that the line following the Start Area has been placed for convenience.  Your line must sit at 90 degrees to your start area, as shown in correct examples 1 & 2.


### No Symmetry

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/no_symmetry.png?raw=true"/>

Your  map must not have any lines of symmetry (e.g., the map cannot be the same route when travelled forwards and backwards). 


### No Bunching

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/no_bunching.png?raw=true"/>

You may not reduce straight line segments below a minimum length of 20cm (including reducing them to 0cm) in order to shorten the overall length of your map ("bunching").

### No Back-Tracking, No Shared Line Start

<img src="https://github.com/paulodowd/EMATM0054_20_21/blob/master/images/no_backtracking.png?raw=true"/>

- The line map you create cannot go behind the start area.  
- Your two start areas cannot lead the Romi on to the same start segment of the line map.
